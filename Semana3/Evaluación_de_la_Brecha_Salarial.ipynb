# **Cálculo de la Brecha Salarial**
## Introducción
La brecha salarial es una medida que evalúa la desigualdad en los ingresos entre diferentes grupos,
como género, etnia, antigüedad o departamento. En términos simples, permite entender qué tan equitativa
es la distribución de salarios dentro de una organización.

El cálculo de la brecha salarial es especialmente relevante para abordar temas de diversidad e inclusión
en el lugar de trabajo. Por ejemplo, una brecha salarial de género significativa puede indicar
la necesidad de implementar políticas más equitativas.

En esta sección, analizaremos:
- Brechas salariales por género.
- Brechas salariales por etnia.
- Brechas combinadas por género y departamento.

Para realizar el cálculo, utilizaremos la fórmula:

image.png

### Brecha Salarial por Género
Analizaremos cómo varían los salarios promedio entre hombres y mujeres, calculando la brecha porcentual
que existe entre ambos grupos.

```python
brecha_genero = df.groupby('Género')['Salario_Anual'].mean()
brecha_genero_calculo = (brecha_genero['Masculino'] - brecha_genero['Femenino']) / brecha_genero['Masculino'] * 100
print(f"Brecha Salarial entre Hombres y Mujeres: {brecha_genero_calculo:.2f}%")
```

### Brecha Salarial por Etnia
Este análisis evalúa la desigualdad en los salarios promedio entre las diferentes etnias presentes
en la organización. Calcularemos la diferencia relativa entre el grupo mejor pagado y el menos pagado.

```python
brecha_etnia = df.groupby('Etnia')['Salario_Anual'].mean()
brecha_etnia_calculo = (brecha_etnia.max() - brecha_etnia.min()) / brecha_etnia.max() * 100
print(f"Brecha Salarial entre las etnias con mayor y menor promedio: {brecha_etnia_calculo:.2f}%")
```

### Brecha Combinada por Género y Departamento
Aquí combinamos dos factores clave: el departamento y el género. Esto nos permitirá observar
si ciertos departamentos muestran mayores desigualdades salariales entre hombres y mujeres.

```python
brecha_comb = df.groupby(['Departamento', 'Género'])['Salario_Anual'].mean().unstack()
brecha_comb['Brecha'] = (brecha_comb['Masculino'] - brecha_comb['Femenino']) / brecha_comb['Masculino'] * 100
print("Brecha Salarial por Departamento:")
print(brecha_comb[['Brecha']])
```

# Visualización de la Brecha Salarial

## Visualizaciones
Las siguientes gráficas permiten interpretar de manera visual las desigualdades detectadas:
- Gráfico de barras para analizar las brechas salariales por género y por etnia.
- Heatmap para observar las brechas combinadas por género y departamento.

```python
# Gráfico de barras para brecha por género
plt.figure(figsize=(8, 6))
brecha_genero.plot(kind='bar', color=['blue', 'pink'])
plt.title("Brecha Salarial por Género", fontsize=16)
plt.xlabel("Género", fontsize=12)
plt.ylabel("Salario Promedio (USD)", fontsize=12)
plt.show()
```

```python
# Gráfico de barras para brecha por etnia
plt.figure(figsize=(10, 6))
brecha_etnia.sort_values().plot(kind='bar', color='green')
plt.title("Brecha Salarial por Etnia", fontsize=16)
plt.xlabel("Etnia", fontsize=12)
plt.ylabel("Salario Promedio (USD)", fontsize=12)
plt.show()
```

```python
# Heatmap para brecha combinada por género y departamento
plt.figure(figsize=(12, 8))
sns.heatmap(brecha_comb[['Brecha']].dropna(), annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'label': 'Brecha (%)'})
plt.title("Brecha Salarial por Género y Departamento", fontsize=16)
plt.xlabel("", fontsize=12)
plt.ylabel("Departamento", fontsize=12)
plt.show()
```
